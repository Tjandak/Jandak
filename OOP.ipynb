## OOP - Objektově orientované programování

OOP je způsob myšlení, kde program modeluje reálný svět. Místo volných proměnných máme jeden objekt, který si svá data nese sebou.

Třídu (= `class`) se hodí napsat, když plánujeme mít ve svém programu více objektů s podobným chováním. Třeba karetní hra by mohla mít třídu *Karta*, webová aplikace třídu *Uzivatel*, tabulkový procesor třídu *Radek*.

**Strukturované programování**

Píšeme kód "shora dolů" - funkce, proměnné, všechno oddělené.

Příklad: vzdálenost bodu od počátku

```python
import math
souradnice_x = [0,1,2,3]
souradnice_y = [0,1,0,4]
seznam_bodu = list(zip(souradnice_x,souradnice_y))
print(f"Seznam bodů: {seznam_bodu}")

def vzdalenost(x,y):
    return math.sqrt(x**2 + y**2)

vysledky = [vzdalenost(souradnice_x[i], souradnice_y[i])for i in range(3)]

# vypis na dve des. mista
for v in vysledky:
    print(f"{v:.2f}", end = " ")
print()
print(f"Vzdalenosti: {vysledky:}")
```

**Objektové programování**

Vytvoříme třídu a z ní děláme konkrétní objekty = *instance*.

Objekt si můžeme představit balíček, který v sobě drží:

* Data (*atributy*) - např. jmeno, vek
* Funkce (*metody*) - co s těmi daty umí dělat

Třída je jako  technický výkres v automobilce - blueprint, podle kterého se vyrábějí auta. Samotná třída žádné auto nemá, to musíme nejprve podle nákresu vyrobit.

Zjednodušeně řečeno - *metoda* je *funkce* definovaná uvnitř třídy.

```python
class Bod:
    def vzdalenost(self):
        return math.sqrt(self.x**2 + self.y**2)

# zatím jsme nevytvořili žádný bod, jen jsme pro něj definovali metodu
vrchol = Bod()
vrchol.x = 1  # Atributy bodu
vrchol.y = 1
vrchol.vzdalenost()
```

**Příklad: Koťátka**

Chceme napsat program o zvířátkách. Začneme tím, že napíšeme třídu pro koťátka, která umí mňoukat:

Tak jako se funkce definují pomocí `def`, třídy mají klíčové slovo `class`, za které napíšeme jméno třídy, dvojtečku a pak odsazené tělo třídy. Podobně jako `def` dělá funkce, příkaz `class` udělá novou třídu a přiřadí ji do proměnné daného jména (tady Kotatko).

Třídy se tradičně pojmenovávají s velkým písmenem (CamelCase), aby se nepletly s „normálními“ hodnotami.

V těle třídy můžeme definovat metody, které vypadají úplně jako funkce – jen mají první parametr `self`.

 Kotatko (s velkým K) je třída – popis, jak se koťátka chovají. mourek (s malým m) je konkrétní objekt (angl. instance) té třídy: hodnota, která reprezentuje kotě.

```python
class Kotatko:
    def zamnoukej(self):
        print("Mňau!")

# Vytvoření konkrétního objektu = instance tridy Kotatko
mourek = Kotatko()

# Volání metody
mourek.zamnoukej()
```

Když definuješ třídu (pomocí bloku class), neznamená to zatím, že ve tvém programu je nějaké koťátko. Třída je jako recept nebo manuál: když si koupíš kuchařku, budeš teoreticky vědět jak upéct dort, jak bude takový dort vypadat a že se dá sníst. Ale neznamená to ještě, že máš samotný dort!

### Atributy

informace, které se uloží k danému objektu. Podle atributů jako jmeno pak můžeš jednotlivá koťátka rozlišit.

```python
mourek = Kotatko()
mourek.jmeno = 'Mourek'

micka = Kotatko()
micka.jmeno = 'Micka'

print(mourek.jmeno)
print(micka.jmeno)
```

### Parametr self
Každá metoda má přístup ke konkrétnímu objektu, na kterém pracuje, právě přes parametr self. Teď, když máš koťátka pojmenovaná, můžeš v metodě zamnoukej použít self a dostat se tak ke jménu daného koťátka:

```python
class Kotatko:
    def zamnoukej(self):
        print(f"{self.jmeno}: Mňau!") ###### pridano {self.jmeno}

mourek = Kotatko()
mourek.jmeno = 'Mourek'

micka = Kotatko()
micka.jmeno = 'Micka'

mourek.zamnoukej()
micka.zamnoukej()
```

## Úkol - vymyslete vlastní třídu
Třída je jako technický výkres v továrně na lokomotivy. Podle jednoho výkresu můžeme vyrobit desítky skutečných strojů.

Tady říkáme, co každá lokomotiva musí mít a co umí.

```python
class Lokomotiva:
    def __init__(self, cislo, typ, max_rychlost):
        self.cislo = cislo
        self.typ = typ
        self.max_rychlost = max_rychlost
        self.aktualni_rychlost = 0
        print(f"✅ Stroj {self.cislo} ({self.typ}) byl vyroben a připraven k jízdě.")

    def zrychli(self, o_kolik):
        nova_rychlost = self.aktualni_rychlost + o_kolik
        if nova_rychlost <= self.max_rychlost:
            self.aktualni_rychlost = nova_rychlost
            print(f"📈 {self.typ} zrychluje na {self.aktualni_rychlost} km/h.")
        else:
            self.aktualni_rychlost = self.max_rychlost
            print(f"⚠️ Pozor! Dosáhli jsme konstrukční rychlosti {self.max_rychlost} km/h!")

    def houkej(self):
            print(f"📢 Lokomotiva {self.cislo} houká: TÚÚÚÚÚÚÚT!")


# Vytvoříme dvě konkrétní parní mašinky
albatros = Lokomotiva("498.1", "Albatros", 120)
rosnicka = Lokomotiva("464.2", "Rosnička", 100)

# Vyzkoušíme jejich funkce
print("-" * 30)
albatros.zrychli(50)
rosnicka.zrychli(160) # Tady by to mělo vyhodit varování o max. rychlosti
rosnicka.houkej()
```

**Vlastní třída**

```python
class Auto:
    def __init__(self, znacka, model, max_rychlost):
        self.znacka = znacka
        self.model = model
        self.max_rychlost = max_rychlost
        self.aktualni_rychlost = 0
        print(f"🚗 Auto {self.znacka} {self.model} je připraveno k jízdě.")

    def zrychli(self, o_kolik):
        nova = self.aktualni_rychlost + o_kolik
        if nova <= self.max_rychlost:
            self.aktualni_rychlost = nova
            print(f"📈 {self.znacka} {self.model} jede {self.aktualni_rychlost} km/h.")
        else:
            self.aktualni_rychlost = self.max_rychlost
            print(f"⚠️ Max rychlost {self.max_rychlost} km/h dosažena!")

    def brzdi(self, o_kolik):
        self.aktualni_rychlost = max(0, self.aktualni_rychlost - o_kolik)
        print(f"🛑 {self.znacka} {self.model} zpomalilo na {self.aktualni_rychlost} km/h.")

    def troubi(self):
        print(f"📢 {self.znacka} {self.model} troubí: PÍP PÍP!")


auto1 = Auto("Škoda", "Octavia", 200)
auto2 = Auto("Tesla", "Model S", 250)

print("-" * 30)
auto1.zrychli(60)
auto1.brzdi(20)

auto2.zrychli(300)
auto2.troubi()
```

# Tři pilíře OOP - dědičnost, polymorfismus a zapouzdření

 **Dědičnost**

`class Pes(Zvire)`: znamená: "*Pes* dědí od *Zvire*"

* Pes a Kočka automaticky dostanou __init__, info(), spat() od rodiče
* Nemusíš psát stejný kód dvakrát
* Každý potomek má svoje speciální schopnosti navíc

**Polymorfismus**

stejný příkaz, jiná reakce - nemusíme zjišťovat typ objektu. Zavoláme metodu a každý objekt ví, jak reagovat.


```python
class Zvire:
    def __init__(self, jmeno):
        self.jmeno = jmeno
    def info(self):
        print(f"Jsem {self.jmeno}, je mi {self.vek} let")
    def spat(self):
        print(f"{self.jmeno} spí... Zzz")
    def ozvi_se(self): # ukazeme na nem polymorfismus
        print(f"{self.jmeno} vydává zvuk")


class Pes(Zvire):
    def ozvi_se(self):  # Přepisujeme metodu rodiče!
        print(f"{self.jmeno}: Haf haf!")
    def aport(self):
        print(f"{self.jmeno} přináší míček")

class Kocka(Zvire):
    def ozvi_se(self):  # Stejná metoda, jiné chování
        print(f"{self.jmeno}: Mňau!")


class Krava(Zvire):
    def ozvi_se(self):
        print(f"{self.jmeno}: Bůůů!")


# 🎯 POLYMORFISMUS V AKCI
zvirata = [
    Pes("Alík"),
    Kocka("Micka"),
    Krava("Květa"),
    Pes("Rex")
]

# Zavoláme STEJNOU metodu na různých objektech
for zvire in zvirata:
    zvire.ozvi_se()  # Každý se ozve PO SVÉM!
```

**Vlastní Zvířata**

```python
class Zvire:
    def __init__(self, jmeno, vek):
        self.jmeno = jmeno
        self.vek = vek

    def info(self):
        print(f"Jsem {self.jmeno} a je mi {self.vek} let")

    def spat(self):
        print(f"{self.jmeno} spí... Zzz")

    def ozvi_se(self):
        print(f"{self.jmeno} vydává zvuk")


class Papousek(Zvire):
    def ozvi_se(self):
        print(f"{self.jmeno}: KŘI KŘI!")

    def opakuj_slova(self, slovo):
        print(f"{self.jmeno} opakuje: {slovo}")


kiki = Papousek("Kiki", 3)

kiki.info()
kiki.spat()
kiki.ozvi_se()
kiki.opakuj_slova("Ahoj!")
```

```python
class Zvire:
    def __init__(self, jmeno):
        self.jmeno = jmeno

    def info(self):
        print(f"Jsem {self.jmeno}")

    def spat(self):
        print(f"{self.jmeno} spí... Zzz")

    def ozvi_se(self):
        print(f"{self.jmeno} vydává zvuk")


class Papousek(Zvire):
    def ozvi_se(self):
        print(f"{self.jmeno}: KŘI KŘI!")

    def opakuj_slova(self, slovo):
        print(f"{self.jmeno} opakuje: {slovo}")


class Chameleon(Zvire):
    def ozvi_se(self):
        print(f"{self.jmeno}: ... (ticho, ale sleduje tě)")

    def zmen_barvu(self, barva):
        print(f"{self.jmeno} mění barvu na {barva}")


class Had(Zvire):
    def ozvi_se(self):
        print(f"{self.jmeno}: Syčí: Ssssss!")

    def plazit_se(self):
        print(f"{self.jmeno} se plazí tiše po zemi")


exoticka_zvirata = [
    Papousek("Kiki"),
    Chameleon("Charlie"),
    Had("Nagini"),
    Papousek("Rio")
]

for zvire in exoticka_zvirata:
    zvire.ozvi_se()
```

**Zapouzdření (encapsulation)**

= skrytí vnitřní implementace a dat.  Objekt si "střeží" své důležité údaje a ostatní s ním můžou pracovat jen přes povolené metody – jako u mobilu: nevidíš, jak funguje uvnitř, ale můžeš používat tlačítka a displej.

**Přístupové modifikátory** určují, kdo může vidět a používat proměnné a metody třídy:

* Public (veřejné) - každý má přístup, píše se normálně: `self.jmeno`
* Private (soukromé) - přístup jen uvnitř třídy, píše se s __: `self.__heslo`
* Protected (chráněné) - přístup pro třídu a její potomky, píše se s _: `self._vek`



```python
class BankovniUcet:
    def __init__(self):
        self.__zustatek = 0  # Skrytý údaj (dvě podtržítka)

    def vlozit(self, castka):
        self.__zustatek += castka

    def zobrazit_zustatek(self):
        return self.__zustatek

ucet = BankovniUcet()
ucet.vlozit(500)
print(ucet.zobrazit_zustatek())  # Funguje: 500
# print(ucet.__zustatek)  # Nefunguje - je to skryté!
```

```python
class Uzivatel:
    def __init__(self):
        self.jmeno = "Petr"           # Public - vidí všichni
        self._vek = 15                # Protected - pro třídu a potomky
        self.__heslo = "tajne123"     # Private - jen uvnitř třídy

    def zobraz_heslo(self):
        return self.__heslo           # Uvnitř třídy funguje

user = Uzivatel()
print(user.jmeno)           # Funguje: Petr
print(user._vek)            # Funguje: 15 (ale nemělo by se)
# print(user.__heslo)       # Nefunguje - je soukromé!
print(user.zobraz_heslo())  # Funguje: tajne123
```

Co by se stalo, kdybychom místo volání friend = Person('Tom', 23) použili pouze friend = Person('Tom')?



```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def describe(self):
        print(self.name, 'is', self.age, 'years old.')

    def rename(self, new_name):
        self.name = new_name

friend = Person('Tom', 23)
friend.rename('David')
friend.describe()
```
