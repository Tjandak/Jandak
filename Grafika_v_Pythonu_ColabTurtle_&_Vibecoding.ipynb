**SPIRÁLA**

```python
from ColabTurtle.Turtle import *
import math

initializeTurtle(initial_speed=13, initial_window_size=(1000,500))

bgcolor('black')

colors = ['magenta', 'yellow', 'cyan', 'lime', 'orange']

steps = 200
length_factor = 3

for i in range(steps):
    color(colors[i % len(colors)])
    width(1 + int(4 * (1 + math.sin(i * 0.3))))
    forward(i * length_factor)
    left(59)
```

**FRAKTÁL**

```python
from ColabTurtle.Turtle import *
import random

initializeTurtle(initial_speed=13, initial_window_size=(1000, 700))
bgcolor('black')
hideturtle()

colors = ["#ff007f", "#ffea00", "#00f5ff", "#00ff73", "#ff6a00"]

def fractal_tree(branch_length, pensize_value):
    if branch_length < 10:
        return
    color(random.choice(colors))
    width(pensize_value)
    forward(branch_length)
    left(30)
    fractal_tree(branch_length * 0.7, max(1, pensize_value-1))
    right(60)
    fractal_tree(branch_length * 0.7, max(1, pensize_value-1))
    left(30)
    backward(branch_length)

penup()
goto(500, 50)
pendown()
setheading(90)

fractal_tree(120, 8)

```

**GALAXIE**

```python
from ColabTurtle.Turtle import *
import random, math

initializeTurtle(initial_speed=13, initial_window_size=(1000, 600))
bgcolor("black")
pensize(4)
showturtle()

def dot(size):
    forward(size)
    backward(size)

for _ in range(80):
    penup()
    goto(random.randint(0, 1000), random.randint(0, 600))
    pendown()
    pencolor("white")
    dot(1)

num_arms = 5
num_points = 80
center_x, center_y = 500, 300

for arm in range(num_arms):
    for i in range(num_points):
        angle = i * 4 + arm * (360 / num_arms)
        radius = i * 2

        x = center_x + radius * math.cos(math.radians(angle))
        y = center_y + radius * math.sin(math.radians(angle))

        penup()
        goto(x, y)
        pendown()

        if i % 10 == 0:
            pencolor(random.choice(["#ff007f", "#00f5ff", "#ffea00", "#00ff73", "#ff6a00"]))
            dot(3)
        elif i % 3 == 0:
            pencolor("white")
            dot(1)

penup()
goto(center_x, center_y)
pendown()
for _ in range(40):
    pencolor(random.choice(["#ffea00", "#ff6a00", "#ff007f", "#00f5ff"]))
    dot(4)
```

**GALAXIE 2**

```python
from ColabTurtle.Turtle import *
import random, math

def draw_gear(radius, teeth, color="gray"):
    pencolor(color)
    fillcolor(color)
    pensize(2)
    penup()
    goto(0, -radius)
    pendown()
    begin_fill()
    for i in range(teeth * 2):
        angle = 360 / (teeth * 2)
        if i % 2 == 0:
            forward_length = radius * 0.15
        else:
            forward_length = radius * 0.3
        forward(forward_length)
        left(angle)
    end_fill()
    penup()
    home()

def draw_galaxy_with_gear(center_x=500, center_y=300):
    initializeTurtle()
    bgcolor("black")
    speed(13)
    pensize(2)
    hideturtle()

    galaxy_colors = [
        "#ff0000", "#cc0000", "#990000",  # červené odstíny
        "#800080", "#993399", "#cc66cc",  # fialové odstíny
        "#0000ff", "#0033cc", "#3399ff",  # modré odstíny
        "#ffffff"                         # bílá
    ]
    ray_colors = ["#ffffff", "#cc66cc", "#3399ff", "#ff0000"]

    for _ in range(400):
        penup()
        goto(random.randint(0, 1000), random.randint(0, 600))
        pendown()
        pencolor("white")
        dot_size = random.uniform(0.2, 0.5)
        forward(dot_size)
        backward(dot_size)

    num_arms = 5
    num_points = 150

    for arm in range(num_arms):
        for i in range(num_points):
            angle = i * 4 + arm * (360 / num_arms)
            radius = i * 2
            x = center_x + radius * math.cos(math.radians(angle))
            y = center_y + radius * math.sin(math.radians(angle))
            penup()
            goto(x, y)
            pendown()
            if i % 5 == 0:
                pencolor(galaxy_colors[i % len(galaxy_colors)])
                forward(random.uniform(0.5, 1.0))
                backward(random.uniform(0.5, 1.0))

    for _ in range(50):
        angle = random.randint(0, 360)
        length = random.randint(50, 200)
        penup()
        goto(center_x, center_y)
        setheading(angle)
        pendown()
        pencolor(random.choice(ray_colors))
        forward(length)
        penup()
        goto(center_x, center_y)

    penup()
    goto(center_x, center_y)
    pendown()
    setheading(0)
    pensize(3)
    draw_gear(50, 20, color="#555555")

draw_galaxy_with_gear()

```
