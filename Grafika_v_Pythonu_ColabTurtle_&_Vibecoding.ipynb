**SPIRÁLA**

```python
from ColabTurtle.Turtle import *
import math

initializeTurtle(initial_speed=13, initial_window_size=(1000,500))

bgcolor('black')

colors = ['magenta', 'yellow', 'cyan', 'lime', 'orange']

steps = 200
length_factor = 3

for i in range(steps):
    color(colors[i % len(colors)])
    width(1 + int(4 * (1 + math.sin(i * 0.3))))
    forward(i * length_factor)
    left(59)
```

**FRAKTÁL**

```python
from ColabTurtle.Turtle import *
import random

initializeTurtle(initial_speed=13, initial_window_size=(1000, 700))
bgcolor('black')
hideturtle()

colors = ["#ff007f", "#ffea00", "#00f5ff", "#00ff73", "#ff6a00"]

def fractal_tree(branch_length, pensize_value):
    if branch_length < 10:
        return
    color(random.choice(colors))
    width(pensize_value)
    forward(branch_length)
    left(30)
    fractal_tree(branch_length * 0.7, max(1, pensize_value-1))
    right(60)
    fractal_tree(branch_length * 0.7, max(1, pensize_value-1))
    left(30)
    backward(branch_length)

penup()
goto(500, 50)
pendown()
setheading(90)

fractal_tree(120, 8)

```

**GALAXIE**

```python
from ColabTurtle.Turtle import *
import random, math

initializeTurtle(initial_speed=13, initial_window_size=(1000, 600))
bgcolor("black")
pensize(4)
showturtle()

def dot(size):
    forward(size)
    backward(size)

for _ in range(80):
    penup()
    goto(random.randint(0, 1000), random.randint(0, 600))
    pendown()
    pencolor("white")
    dot(1)

num_arms = 5
num_points = 80
center_x, center_y = 500, 300

for arm in range(num_arms):
    for i in range(num_points):
        angle = i * 4 + arm * (360 / num_arms)
        radius = i * 2

        x = center_x + radius * math.cos(math.radians(angle))
        y = center_y + radius * math.sin(math.radians(angle))

        penup()
        goto(x, y)
        pendown()

        if i % 10 == 0:
            pencolor(random.choice(["#ff007f", "#00f5ff", "#ffea00", "#00ff73", "#ff6a00"]))
            dot(3)
        elif i % 3 == 0:
            pencolor("white")
            dot(1)

penup()
goto(center_x, center_y)
pendown()
for _ in range(40):
    pencolor(random.choice(["#ffea00", "#ff6a00", "#ff007f", "#00f5ff"]))
    dot(4)
```

**GALAXIE 2**

```python
import matplotlib.pyplot as plt
import math
import random
import numpy as np

x = []
y = []
sizes = []
colors = []

arms = 5
stars_per_arm = 800

for arm in range(arms):
    offset = 2 * math.pi * arm / arms
    for i in range(stars_per_arm):
        angle = i * 0.35 + offset
        radius = 0.45 * i

        spread = random.gauss(0, 10)
        r = radius + spread

        x.append(r * math.cos(angle))
        y.append(r * math.sin(angle))

        sizes.append(random.uniform(0.5, 3))

        color = plt.cm.plasma(i / stars_per_arm)
        colors.append(color)

bg_x = np.random.uniform(-600, 600, 600)
bg_y = np.random.uniform(-600, 600, 600)

plt.figure(figsize=(9, 9), facecolor="black")
plt.scatter(bg_x, bg_y, s=1, c="white", alpha=0.4)
plt.scatter(x, y, s=sizes, c=colors, alpha=0.9)

plt.gca().set_facecolor("black")
plt.axis("off")
plt.show()
```
