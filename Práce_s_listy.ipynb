```python
import matplotlib.pyplot as plt
import statistics

tramvaje = [
    ("T2", 60, 1955, 9.5),
    ("T3", 65, 1960, 11.2),
    ("T6A5", 60, 1985, 10.0),
    ("KT8D5", 70, 1986, 14.3),
    ("Škoda 13T", 75, 2002, 12.5),
    ("Škoda 14T", 80, 2006, 13.0),
    ("Škoda 15T", 80, 2008, 12.8),
    ("Škoda 16T", 85, 2010, 14.0),
    ("Pragoimex VarioLF", 75, 2003, 11.5),
    ("Škoda 19T", 90, 2018, 15.0)
]

rychlejsi_tramvaje = [t for t in tramvaje if t[1] > 70]
serazene_podle_roku = sorted(tramvaje, key=lambda x: x[2])
serazene_podle_rychlosti = sorted(tramvaje, key=lambda x: x[1], reverse=True)
serazene_podle_delky = sorted(tramvaje, key=lambda x: x[3], reverse=True)

rychlosti = [t[1] for t in tramvaje]
delky = [t[3] for t in tramvaje]
prumer_rychlost = statistics.mean(rychlosti)
mediana_rychlost = statistics.median(rychlosti)
rozptyl_rychlost = statistics.variance(rychlosti)
min_rychlost = min(rychlosti)
max_rychlost = max(rychlosti)

prumer_delka = statistics.mean(delky)
mediana_delka = statistics.median(delky)
rozptyl_delka = statistics.variance(delky)
min_delka = min(delky)
max_delka = max(delky)

typy_tramvaji = [t[0] for t in tramvaje]
roky = [t[2] for t in tramvaje]

print("Tramvaje s max. rychlostí > 70 km/h:", rychlejsi_tramvaje)
print("\nSeřazené podle roku výroby:", serazene_podle_roku)
print("\nSeřazené podle max. rychlosti:", serazene_podle_rychlosti)
print("\nSeřazené podle délky tratě:", serazene_podle_delky)

print(f"\nRychlosti - průměr: {prumer_rychlost}, medián: {mediana_rychlost}, min: {min_rychlost}, max: {max_rychlost}, rozptyl: {rozptyl_rychlost:.2f}")
print(f"Délky tratí - průměr: {prumer_delka}, medián: {mediana_delka}, min: {min_delka}, max: {max_delka}, rozptyl: {rozptyl_delka:.2f}")

plt.figure(figsize=(8,5))
plt.hist(rychlosti, bins=6, edgecolor='black', color='skyblue')
plt.title("Histogram maximálních rychlostí tramvají")
plt.xlabel("Rychlost [km/h]")
plt.ylabel("Počet tramvají")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

plt.figure(figsize=(10,5))
plt.bar(typy_tramvaji, rychlosti, color='orange', edgecolor='black')
plt.title("Maximální rychlost jednotlivých typů tramvají")
plt.xlabel("Typ tramvaje")
plt.ylabel("Maximální rychlost [km/h]")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

plt.figure(figsize=(10,5))
plt.bar(typy_tramvaji, delky, color='green', edgecolor='black')
plt.title("Délka tratě jednotlivých typů tramvají")
plt.xlabel("Typ tramvaje")
plt.ylabel("Délka tratě [km]")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

plt.figure(figsize=(8,5))
plt.scatter(roky, rychlosti, color='red')
plt.title("Rok výroby vs maximální rychlost tramvají")
plt.xlabel("Rok výroby")
plt.ylabel("Maximální rychlost [km/h]")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()

plt.figure(figsize=(8,5))
plt.scatter(roky, delky, color='purple')
plt.title("Rok výroby vs délka tratě tramvají")
plt.xlabel("Rok výroby")
plt.ylabel("Délka tratě [km]")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
