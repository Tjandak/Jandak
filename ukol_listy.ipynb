```python
asistenti = ["ChatGPT", "Claude", "Bard", "Llama"]

print(asistenti)
print("-" * 40)

nova_polozka = "Perplexity"
asistenti.append(nova_polozka)
print(asistenti)
print("-" * 40)

if asistenti:
    asistenti.pop(0)
print(asistenti)
print("-" * 40)

print(len(asistenti))
print("-" * 40)

for index, polozka in enumerate(asistenti, start=1):
    print(index, polozka)
```
