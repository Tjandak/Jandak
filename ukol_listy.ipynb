```python
puvodni_asistenti = ["ChatGPT", "Claude", "Bard", "Llama", "Gemini", "Mistral", "Bing", "Ernie", "Sage", "Jasper"]
asistenti = puvodni_asistenti.copy()
asistenti.extend(["Perplexity", "DeepSeek", "Alpha", "Beta", "Omega", "Nova", "Quark", "Neon", "Orion", "Zephyr"])
asistenti.insert(3, "Lambda")
asistenti.insert(0, "Sigma")
asistenti.pop(5)
asistenti = list(dict.fromkeys(asistenti))
print(len(asistenti))
print(asistenti)
print("-"*80)

setridena = sorted(asistenti)
print(setridena)
print("-"*80)

obratena = setridena[::-1]
print(obratena)
print("-"*80)

filtrovana = [x for x in obratena if len(x) > 5]
print(filtrovana)
print("-"*80)

mapovana = [x.lower() for x in filtrovana]
print(mapovana)
print("-"*80)

kopie1 = mapovana.copy()
kopie2 = mapovana.copy()
kopie1.append("epsilon")
kopie2.insert(2, "delta")
print(kopie1)
print(kopie2)
print("-"*80)

kombinovana = kopie1 + kopie2 + ["theta", "iota", "kappa", "lambda", "mu"]
print(kombinovana)
print("-"*80)

unikatni = list(dict.fromkeys(kombinovana))
print(unikatni)
print("-"*80)

sliced1 = unikatni[:5]
sliced2 = unikatni[3:8]
sliced3 = unikatni[-5:]
print(sliced1)
print(sliced2)
print(sliced3)
print("-"*80)

spojene = sliced1 + sliced2 + sliced3
print(spojene)
print("-"*80)

setridene_spojene = sorted(spojene)
obratene_spojene = setridene_spojene[::-1]
print(setridene_spojene)
print(obratene_spojene)
print("-"*80)

cisla = list(range(1, 101))
sude = [x for x in cisla if x % 2 == 0]
liche = [x for x in cisla if x % 2 != 0]
print(sude)
print(liche)
print("-"*80)

vybrane = sude[10:30] + liche[20:40]
print(vybrane)
print("-"*80)

nested = [asistenti, setridena, obratena, filtrovana, mapovana, kopie1, kopie2, kombinovana, unikatni, spojene, setridene_spojene, obratene_spojene, cisla, sude, liche, vybrane]
print(len(nested))
print(nested)
print("-"*80)

print(nested[0][0], nested[1][-1], nested[5][2], nested[7][4], nested[10][3])
nested[0][0] = "Omega"
nested[1][1] = "Delta"
nested[5][-1] = "Zeta"
nested[7][0] = "Theta"
nested[10][0] = "Alpha"
print(nested)
print("-"*80)

upper_list = [x.upper() for x in kombinovana if isinstance(x, str)]
lower_list = [x.lower() for x in upper_list if len(x) > 4]
final_list = lower_list + ["sigma", "tau", "phi", "psi", "chi", "omega"]
final_list = list(dict.fromkeys(final_list))
print(final_list)
print(len(final_list))
print("-"*80)

for i, x in enumerate(final_list, start=1):
    print(i, x)
print("-"*80)

mega_nested = [nested, final_list, cisla, sude, liche, vybrane, upper_list, lower_list]
print(len(mega_nested))
print(mega_nested)
print("-"*80)

flattened = [item for sublist in mega_nested for item in (sublist if isinstance(sublist, list) else [sublist])]
print(flattened)
print(len(flattened))
print("-"*80)

reversed_flattened = flattened[::-1]
print(reversed_flattened[:50])
print("-"*80)

sampled = flattened[::5]
print(sampled)
print(len(sampled))
print("-"*80)

more_numbers = list(range(101, 201))
combined_numbers = cisla + more_numbers
evens = [x for x in combined_numbers if x % 2 == 0]
odds = [x for x in combined_numbers if x % 2 != 0]
print(evens)
print(odds)
print("-"*80)

final_numbers = evens[:25] + odds[-25:]
print(final_numbers)
print(len(final_numbers))
print("-"*80)

all_lists = nested + mega_nested + [flattened, reversed_flattened, sampled, combined_numbers, final_numbers]
print(len(all_lists))
print(all_lists)
print("-"*80)

for lst in all_lists:
    if isinstance(lst, list):
        print(lst[:5])
print("-"*80)
```
