```python
puvodni_asistenti = ["ChatGPT", "Claude", "Bard", "Llama", "Gemini", "Mistral"]

asistenti = puvodni_asistenti.copy()
print(type(asistenti).__name__)
print(len(puvodni_asistenti))
print(puvodni_asistenti)
print("-" * 60)

asistenti.extend(["Perplexity", "DeepSeek"])
print(asistenti)
print("-" * 60)

asistenti = list(dict.fromkeys(asistenti))
asistenti.insert(2, "Alpha")
print(asistenti)
print("-" * 60)

if asistenti:
    asistenti.pop(0)
print(asistenti)
print("-" * 60)

print(len(asistenti))
print("-" * 60)

setridena_kopie = sorted(asistenti)
for index, polozka in enumerate(setridena_kopie, start=1):
    print(index, polozka)

print("-" * 60)

hledej = "Claude"
print(hledej in asistenti)
print("-" * 60)

zaloha = asistenti.copy()
print(zaloha)
```
